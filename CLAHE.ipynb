#CLAHE
Contrast limited adaptive histogram equalization (CLAHE)

```
from google.colab import drive
drive.mount('/content/drive')
```

```
import matplotlib.pyplot as plt
import cv2
import numpy as np


im = plt.imread('/content/drive/My Drive/lowcontrast.jpg')
plt.imshow(im)
plt.show()
```

```
lab_img= cv2.cvtColor(im, cv2.COLOR_BGR2LAB)
# l stands for Luminosity
l,a,b= cv2.split(lab_img)
plt.hist(l.flat, bins=100, range=(0,255))

```

```
eq_l=cv2.equalizeHist(l)
plt.hist(eq_l.flat, bins=100, range=(0,255))
```

```
plt.imshow(eq_l,cmap='gray')
```

```
updated_img1= cv2.merge((eq_l, a, b))
hist_eq_img= cv2.cvtColor(updated_img1, cv2.COLOR_LAB2BGR)
```

```
clahe= cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
clahe_img= clahe.apply(l)
updated_img2= cv2.merge((clahe_img, a, b))
CLAHE_img= cv2.cvtColor(updated_img2, cv2.COLOR_LAB2BGR)
```

```
cv2.imshow('Original Image', im)
cv2.imshow('Equalized Image', hist_eq_img)
cv2.imshow('CLAHE Image', CLAHE_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
